```python
import pandas as pd

# Données : 5 élèves et 4 matières
data = {
    'Maths': [15, 12, 18, 10, 14],
    'Physique': [14, 11, 17, 9, 13],
    'Français': [8, 12, 6, 14, 10],
    'Histoire': [9, 13, 7, 15, 11]
}

df = pd.DataFrame(data, index=['Élève 1', 'Élève 2', 'Élève 3', 'Élève 4', 'Élève 5'])
print(df)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)  # On réduit à 2 dimensions
X_pca = pca.fit_transform(X_scaled)
```

```python
# Créer un DataFrame avec les composantes principales
df_pca = pd.DataFrame(X_pca, index=df.index, columns=['PC1', 'PC2'])
print(df_pca)
```

```python
# Pourcentage de variance expliquée par chaque composante
print("Variance expliquée :", pca.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
plt.scatter(df_pca['PC1'], df_pca['PC2'])

for i, name in enumerate(df.index):
    plt.text(df_pca['PC1'][i], df_pca['PC2'][i], name)

plt.xlabel('Composante principale 1')
plt.ylabel('Composante principale 2')
plt.title("ACP des élèves")
plt.grid(True)
plt.show()
```
