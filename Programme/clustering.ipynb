```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
dispo_pivot = pd.read_csv("../Source/clean_data/diapo_alimentaire.csv", sep=",", index_col="Zone")
dispo_pivot
```

```python
scaler = StandardScaler()
dispo_pivot_scaled = scaler.fit_transform(dispo_pivot)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(dispo_pivot_scaled)
```

```python
print("Variance expliquée par chaque composante :", pca.explained_variance_ratio_)
```

```python
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel('Première Composante Principale')
plt.ylabel('Deuxième Composante Principale')
plt.title('ACP')
plt.show()
```

```python
print(pca.explained_variance_ratio_)
```

## ACP 3dimensions

```python
dispo_pivot.head()
```

```python
# Standardisation
X_std = StandardScaler().fit_transform(dispo_pivot)
```

```python

# ACP sur 3 composantes
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_std)
```

```python

# Création d’un DataFrame avec les 3 composantes
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3'])
```

```python

# 🎨 Visualisation 3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
```

```python

# Tu peux ajouter une couleur selon une variable catégorielle si tu veux
ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], color='blue', s=10, alpha=0.6)
```

```python

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title("Nuage de points après ACP (3D)")
plt.show()
```

```python
# Ajouter les labels (zones)
df_pca['Zone'] = dispo_pivot.index

# Représentation en 3D avec couleur selon les zones
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Générer des couleurs pour chaque point
colors = sns.color_palette("hsv", len(df_pca['Zone'].unique()))
zone_colors = dict(zip(df_pca['Zone'].unique(), colors))
point_colors = df_pca['Zone'].map(zone_colors)

# Tracer le nuage
ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'],
           c=point_colors, s=20, alpha=0.8)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title("Nuage de points après ACP (3D) - par Zone")

plt.show()
```
